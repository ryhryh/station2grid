```python
result = pd.DataFrame()

for x in epa_station_info.SiteEngName[:2]:
    print('-'*80)
    print(x)
    
    opt3.val_stations = x
    
    station2GridSD = station2gridSD_model.Station2GridSD(opt3)
    grids = station2GridSD.test() ###
    
    stations = station2GridSD.stations
    stations = station2GridSD.dataGenerator.denormalize(stations)
    
    df = epa_station_info[epa_station_info.SiteEngName==x]
    indexs = df.index.values
    rows = df.row.values
    cols = df.col.values
    
    dummy = pd.DataFrame()
    dummy['real'] = stations[:,indexs,0].reshape(-1)
    dummy['predict'] = grids[:,rows,cols,0].reshape(-1)
    dummy['station'] = x
    dummy['source'] = station2GridSD.source
    result = result.append(dummy)
```

# temp

```python
def get_oneStation(epa_station_path,val_station,source,grids):
    df = epa_station_info[epa_station_info.SiteEngName==val_station]
    indexs = df.index.values
    rows = df.row.values
    cols = df.col.values

    all_stations = np.load(epa_station_path)

    dummy = pd.DataFrame()
    dummy['real'] = all_stations[:,indexs,2].reshape(-1)
    dummy['predict'] = grids[:,rows,cols,0].reshape(-1)
    dummy['station'] = val_station
    dummy['source'] = source
    year=epa_station_path.split('/')[-1][3:7]
    dts=pd.date_range(start=year, end=str(int(year)+1), periods=None, freq='H', closed='left')
    dummy['dt'] = dts
    return dummy
```

```python
opt=Opt()
epa_station_info = CommonObj().epa_station_info

for k in [1]: #,3,5
    for weightKNN in ['uniform']: #,'distance'
        if (k,weightKNN)==(1,'distance'): continue
            
        print('#'*80)
        print(k,weightKNN)
        
        opt.domain='sat'
        opt.n_epochs=10
        opt.batch_size=300
        opt.k=k
        opt.weightKNN=weightKNN
        
        #--------------------------------------------------------------------------
        #--------------------------------------------------------------------------

        opt.model_name='grid2code'
        grid2code = grid2code_model.Grid2Code(opt)
        grid2code.train()
        grid2code.test()

        #--------------------------------------------------------------------------
        #--------------------------------------------------------------------------
        opt.model_name='station2code'
        opt.features='pm25'
        for x in epa_station_info.SiteEngName[20:21]:
            print('-'*80)
            print(x)

            opt.val_stations = x
            station2code = station2code_model.Station2Code(opt)
            station2code.train()

            
        #--------------------------------------------------------------------------
        #--------------------------------------------------------------------------
        opt.model_name='station2gridSD'
        opt.epa_station_path=os.path.join(home, 'station2grid', 'datasets', 'npy', 'epa', 'epa2019.npy') 

        result = pd.DataFrame()
        for x in epa_station_info.SiteEngName[20:21]:
            print('-'*80)
            print(x)
            opt.val_stations = x
            station2GridSD = station2gridSD_model.Station2GridSD(opt)
            grids = station2GridSD.test() ###
            
            oneStation = get_oneStation(opt.epa_station_path,x,station2GridSD.source,grids)
            result = result.append(oneStation)
#         result.to_csv('%s.csv'%(station2GridSD.source), index=False)
```
