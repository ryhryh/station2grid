```python
from tools import CommonObj, plotMap
from tools.knn_model import KnnModel
from tools.krg_model import KrgModel
from tools.options import Opt
from models import grid2code_model, station2code_model, station2gridSD_model
```

```python
epa_station_info = CommonObj().epa_station_info
epa_station_path = '/media/disk3/feynman52/station2grid/datasets/npy/epa/epa2019.npy'
```

```python
val_station = epa_station_info.SiteEngName.values[-1]
val_station
```

```python
val_stations = [val_station]
model_krg = KrgModel(epa_station_path, val_stations)
grids_krg = model_krg.get_grids()
stations_krg = model_krg.stations
grids_krg.shape, stations_krg.shape
```

```python
val_stations = [val_station]
model_knn = KnnModel(epa_station_path, val_stations)
grids_knn = model_knn.get_grids(4,'distance')
stations_knn = model_knn.stations
grids_knn.shape, stations_knn.shape
```

```python
opt=Opt()
opt.model_name='station2gridSD'
opt.epa_station_path=epa_station_path

opt.domain='air' 
opt.k=3
opt.weightKNN='distance'
opt.features='pm25'
opt.val_stations = val_station

station2GridSD = station2gridSD_model.Station2GridSD(opt)
grids_air = station2GridSD.test() 

grids_air.shape
```

```python
opt=Opt()
opt.model_name='station2gridSD'
opt.epa_station_path=epa_station_path

opt.domain='sat' 
opt.k=3
opt.weightKNN='distance'
opt.features='pm25'
opt.val_stations = val_station

station2GridSD = station2gridSD_model.Station2GridSD(opt)
grids_sat = station2GridSD.test() 

grids_sat.shape
```

```python
i = 3

fig_num = 5
plt.figure(figsize=(fig_num*3,fig_num*2),dpi=100)

###########################################################################
ax=plt.subplot(1, fig_num, 1)
lats = epa_station_info.lat.values
lons = epa_station_info.lon.values
vals = stations_krg[i,:,2].reshape(-1)
title = 'stations_epa'
plotMap(ax,lats,lons,vals,title,fig_num)
###########################################################################
ax=plt.subplot(1, fig_num, 2)
lats = model_knn.coordinates[:,0]
lons = model_knn.coordinates[:,1]
vals = grids_air[i,...].reshape(-1)
title = 'grids_air'
plotMap(ax,lats,lons,vals,title,fig_num)
###########################################################################
ax=plt.subplot(1, fig_num, 3)
lats = model_knn.coordinates[:,0]
lons = model_knn.coordinates[:,1]
vals = grids_sat[i,...].reshape(-1)
title = 'grids_sat'
plotMap(ax,lats,lons,vals,title,fig_num)
###########################################################################
ax=plt.subplot(1, fig_num, 4)
lats = model_knn.coordinates[:,0]
lons = model_knn.coordinates[:,1]
vals = grids_knn[i,...].reshape(-1)
title = 'grids_knn'
plotMap(ax,lats,lons,vals,title,fig_num)
###########################################################################
ax=plt.subplot(1, fig_num, 5)
lats = model_knn.coordinates[:,0]
lons = model_knn.coordinates[:,1]
vals = grids_krg[i,...].reshape(-1)
title = 'grids_krg'
plotMap(ax,lats,lons,vals,title,fig_num)

plt.tight_layout(w_pad=4)
plt.show()
```
