```python
def get_one_dt(station_path):
    pat='station/(.+)_'
    tar=station_path
    dt=pd.to_datetime(re.search(pat,tar).group(1))
    station=np.load(station_path)
    
    if dt.minute==30: return ### remove 10:30 because epa w/o half hour
    include_cols=['pm25', 'pm10', 'temperature', 'humidity', 'PM25', 'PM10',
        'AMBTEMP', 'RH']
    
    one_dt=pd.DataFrame()
    for i_s,s in enumerate(info.SiteEngName):
        for i_f,f in enumerate(domain_features):
            if f not in include_cols: continue ###
            dummy=pd.DataFrame()
            dummy['station']=[s]
            dummy['feature']=[f]
            dummy['dt']=[dt]
            dummy['value']=[station[0,i_s,i_f]]
            one_dt=one_dt.append(dummy)
    return one_dt 

# get_one_dt(station_path)
```

## corre by station

```python
import scipy
def r_squared(x, y):
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x, y)
    return r_value**2

def f(group):
    dummy=pd.DataFrame()
    dummy['pm25'] = [r_squared(group.pm25,group.PM25)]
    dummy['pm10'] = [r_squared(group.pm10,group.PM10)]
    dummy['temperature'] = [r_squared(group.temperature,group.AMBTEMP)]
    dummy['humidity'] = [r_squared(group.humidity,group.RH)]
    return dummy

by=['station']
r2_by_station=pivot.groupby(by).apply(f).reset_index(level=0, drop=False)
r2_by_station.head()
```

```python
r2_by_station_=pd.melt(
    r2_by_station, 
    id_vars=['station'], 
    value_vars=r2_by_station.columns[1:],
    var_name='feature',
    value_name='r_squared',
)

r2_by_station_.head()
```

```python
def get_prob_dist(arr,arr_range=(0,2),bins=10):
    count, division = np.histogram(
        arr,
        bins=bins,
        range=arr_range,
    )    
    pd=count/sum(count)
    return pd

get_prob_dist(df.width, arr_range=(0,2), bins=10)
```

```python
def get_epa_dom(station='Tainan',epa_feature='PM10',dom_feature='pm10'):
    epa_=df_epa[(df_epa.feature==epa_feature)&(df_epa.station==station)]
    dom_=df_dom[(df_dom.feature==dom_feature)&(df_dom.station==station)]
    return epa_,dom_
```

```python
def plot_hist(epa_,dom_,station='Tainan',epa_feature='PM10',dom_feature='pm10',kld=None,x_range=(-10,200)):
    sns.distplot(
        epa_.value, 
        bins=50, 
        hist=False, 
        label='-'.join(['epa',station,epa_feature]),
        norm_hist=True
    
    )

    ax=sns.distplot(
        dom_.value, 
        bins=50, 
        hist=False, 
        label='-'.join(['dom',station,dom_feature]),
        norm_hist=True
    )
    ax.set(title = 'KLD = %.2f'%(kld))
    ax.set_xlim(x_range)
    ax.set_ylim([0,0.15])
    plt.legend()
    plt.show()
```

```python
def get_kld(epa_,dom_):
    epsilon=np.finfo(float).eps
    epa_pd=get_prob_dist(epa_.value, arr_range=(0,350),bins=50)+epsilon
    dom_pd=get_prob_dist(dom_.value, arr_range=(0,350),bins=50)+epsilon

    kld=entropy(pk=dom_pd, qk=epa_pd)
    return kld

```
