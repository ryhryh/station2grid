# add new time label

```python
def change_time_interval(dt,time_interval):
    dt=pd.to_datetime(dt)
    minute = dt.minute
    second = dt.second
    minute = minute+(second/60)
    minute = round(minute/time_interval) * time_interval
    dt = dt.replace(minute=0,second=0)
    dt = dt+pd.Timedelta(minutes=minute)
    return dt
    
0
15
30
45
60

time_interval=30
print(change_time_interval('2018-11-28 22:14:31',time_interval))
print(change_time_interval('2018-11-28 22:16:31',time_interval))
print(change_time_interval('2018-11-28 22:44:31',time_interval))
```

```python
df2['dt']=df2.apply(lambda row: row['Date']+' '+row['Time'],axis=1)
df2.head()
```

```python
df2['dt2']=df2['dt'].apply(lambda x: change_time_interval(x,30))
df2.head()
```

```python
df3_median=df2.groupby(by=['dt2','lat','lon']).median().reset_index(drop=False)
df3_median.head()
```

# rename columns

```python
columns={
    'dt2':'dt',
    'PM2.5':'pm25',
    'PM10':'pm10',
    'Temperature':'temperature',
    'Humidity':'humidity',
}
df3_median=df3_median.rename(columns=columns)
df3_median.head()
```

# script

```python
def process_one_year(year):
    df=pd.read_csv('airbox-%s.csv'%(year))
    df2=df[:100000].copy() ###
    df2['dt']=df2.apply(lambda row: row['Date']+' '+row['Time'],axis=1)
    df2['dt2']=df2['dt'].apply(lambda x: change_time_interval(x,30))
    df3_median=df2.groupby(by=['dt2','lat','lon']).median().reset_index(drop=False)
    df3_median=df3_median.drop(columns=['PM1'])
    columns={
        'dt2':'dt',
        'PM2.5':'pm25',
        'PM10':'pm10',
        'Temperature':'temperature',
        'Humidity':'humidity',
    }
    df3_median=df3_median.rename(columns=columns)
    return df3_median
```

```python
dfs=[process_one_year(year) for year in [2017,2018]]            
df=pd.concat(dfs,axis=0)
df.head()
```
